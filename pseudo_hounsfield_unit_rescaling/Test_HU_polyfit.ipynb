```python
import numpy as np
import matplotlib.pyplot as plt

def calc_polynomial(z, x):
    degree = len(z) - 1
    result = 0
    
    for idx_degree in range(degree+1): 
        result += z[idx_degree] * x ** idx_degree

    return result

def plot_polynomial(z, x_array): 
    result_array = np.zeros_like(x_array)

    for idx_x_array in range(len(x_array)):
        result_array[idx_x_array] = calc_polynomial(z, x_array[idx_x_array])

    fig, ax = plt.subplots(1, 1)
    ax.plot(x_array, result_array)
    # ax.plot(result_array, x_array)

    return fig

known_hu = [-1000, 0]
measured_gv = [5323.07, 6788.4]

z = np.polyfit(known_hu, measured_gv, 1)
z

fig = plot_polynomial(z, np.linspace(-10000, 10000))
fig.show()
```
